Loading of Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading of the Dataset

```python
df = pd.read_csv('climate.csv', skiprows=4)  
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

Visualize the Data:

Line Plot of Temperature Anomalies Over Time:

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Anomaly')
plt.title('Global Temperature Anomalies Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')
plt.show()
```

Histogram of Anomalies:

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Anomaly'], bins=30, kde=True)
plt.title('Distribution of Temperature Anomalies')
plt.xlabel('Temperature Anomaly (°C)')
plt.ylabel('Frequency')
plt.show()
```
